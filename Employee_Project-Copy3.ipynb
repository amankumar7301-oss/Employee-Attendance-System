```python
import pandas as pd


df = pd.read_excel(r'C:\Users\user\Desktop\Employee_Project\Employee_data.xlxs.xlsx',sheet_name='hr')



print(df)
```

```python
hr_present = (df["Status"] == "Present").sum()
hr_absent = (df["Status"] == "Absent").sum()
hr_total = len(df)

print("HR Total:", hr_total)
print("HR Present:", hr_present)
print("HR Absent:", hr_absent)
```

```python
summary = pd.DataFrame({
    "Department": ["IT", "HR"],
    "Total": [total, hr_total],
    "Present": [present, hr_present],
    "Absent": [absent, hr_absent]
})

summary
```

```python
summary.to_excel("summary.xlsx", index=False)
```

```python
import matplotlib.pyplot as plt

summary.set_index("Department")[["Present", "Absent"]].plot(kind="bar")

plt.title("Employee Attendance")
plt.ylabel("Count")
plt.show()
```

```python
def predict_attendance(present_count):
    if present_count >= 3:
        return "Good"
    else:
        return "Poor"

summary["Performance"] = summary["Present"].apply(predict_attendance)

summary
```
